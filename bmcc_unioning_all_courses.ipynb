# Combining Course Information From Every Department
* In this example I will only do the first 4 departments

## Importing Packages

```python
import pandas as pd
```

## Combining CSV Files

```python
df1 = pd.read_csv('accounting.csv')
df2 = pd.read_csv('academic_literacy_and_linguistics.csv')
df3 = pd.read_csv('allied_health.csv')
df4 = pd.read_csv('business_management.csv')
```

```python
all_dfs = [df1, df2, df3, df4]

all_courses = pd.concat(all_dfs, ignore_index=True)
all_courses
```

## Exporting

```python
all_courses.to_csv('BMCC_Courses.csv', index=False, encoding='utf-8')
```
